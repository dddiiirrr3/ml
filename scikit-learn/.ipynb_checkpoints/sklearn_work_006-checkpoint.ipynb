```python
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
df_train = pd.read_csv('../data/home-data-for-ml-course/train.csv')
df_test = pd.read_csv('../data/home-data-for-ml-course/test.csv')
df_sample = pd.read_csv('../data/home-data-for-ml-course/sample_submission.csv')
```

```python
features = ['OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'GrLivArea', 'FullBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'YrSold']
```

```python
y_train = df_train.SalePrice
```

```python
X_train = df_train[features]
```

```python
X_test = df_test[features]
```

```python
y_test = df_sample['SalePrice']
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(X_train, y_train)
    preds_val = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds_val)
    return(mae)
```

```python
maes = []
for max_leaf_nodes in range(3, 1000):
    my_mae = get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test)
    maes = maes + [my_mae]
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
best_tree_size = maes.index(min(maes))+2
print(best_tree_size)
```
