## **map**

```python
# transform a series
points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - points_mean)
```

```python
# transform all dataframe
def remean_points(row):
    row.points = row.points - points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

## **group**

```python
# group by points value
reviews.groupby('points').points.count()
```

```python
# group by count points value
reviews.points.value_counts()
```

```python
# show min price of every points
reviews.groupby('points').price.min()
```

```python
# show mean price of every points
reviews.groupby('points').price.mean()
```

```python
# show max price of every points
reviews.groupby('points').price.max()
```

```python
# sort by 'winery' column and show first wine title in this winery
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# sort by country and show min points wine in country
reviews.groupby(['country']).apply(lambda df: df.loc[df.points.idxmin])
```

```python
# show quantity of wines in every country, min price, mean price, max price
reviews.groupby(['country']).price.agg([len, min, np.mean, max])
```

```python
# grouping by country, show how many wines in every province
reviews.groupby(['country', 'province']).title.agg([len])
```

## **sort**

```python
# sort country-province by len
reviews.groupby(['country', 'province']).title.agg([len]).sort_values(by='len')
```

```python
# sort country-province by len and start from max len
reviews.groupby(['country', 'province']).title.agg([len]).sort_values(by='len', ascending=False)
```

```python
# sorting and adding number index
reviews.groupby(['country', 'province']).title.agg([len]).reset_index()
```

```python
# sorting by country and sorting province in country by len
reviews.groupby(['country', 'province']).title.agg([len]).reset_index().sort_values(by=['country', 'len'])
```
